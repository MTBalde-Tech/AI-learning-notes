
## 5. Plotting a Linear Function  
Linear functions are widely used in machine learning:

\[
y = wx + b
\]


```python

import numpy as np
import matplotlib.pyplot as plt

w = 2
b = 1

x = np.linspace(-10, 10, 200)
y = w * x + b

plt.figure()
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Function: y = 2x + 1")
plt.grid(True)
plt.show()
```
